- VẤN ĐỀ 5: Thu nhập của 2 nhóm khách hàng 

```python
fig, ax = plt.subplots(1,2,figsize=(14,5))

# Left plot
sns.histplot(data=data, x ='Income',hue='Personal Loan', kde=True, ax=ax[0])

# Right plot

sns.violinplot(data=data, y ='Income', x= 'Personal Loan' , ax=ax[1])

# text plot for the people that does not accept the personal loan
ax[1].plot(data.query('Personal Loan == 0')['Income'].median(), 0, 'bo')
ax[1].text(data.query('Personal Loan == 0')['Income'].median() + 0.2,-0.1, f"{data.query('Personal Loan == 0')['Income'].median()}")

# text plot for the people that accept the personal loan
ax[1].plot(data.query('Personal Loan == 1')['Income'].median(), 1, 'bo')
ax[1].text(data.query('Personal Loan == 1')['Income'].median() + 0.2, 0.9, f"{data.query('Personal Loan == 1')['Income'].median()}")


plt.show()
```

```python
data["Income_group"] = pd.cut(
    x=data["Income"],
    bins=[0, 50, 140, 224],
    labels=["Lower", "Middle", "High"])
#tạo ra cột mới 
```

```python
sns.relplot(x='Income_group',y='CCAvg',hue='Personal Loan',data=data)
sns.despine(top=True,right=True,left=True) # to remove side line from graph
```

Trong biểu đồ này, chúng ta có thể thấy rằng chúng ta có sự phân biệt rõ ràng về thu nhập giữa những người có chấp nhận khoản vay cá nhân hay không, những người đã chấp nhận khoản vay có thu nhập cao hơn những người không chấp nhận.

Độ tuổi của khách hàng nằm trong khoảng từ 23 - 67, với giá trị trung bình và trung vị là ~45

```python
sns.distplot(data[data['Personal Loan'] == 0]['Age'], color = 'g')
sns.distplot(data[data['Personal Loan'] == 1]['Age'], color = 'r')
sns.despine(top=True,right=True,left=True)
```

Những người có thu nhập cao hơn đã chọn vay cá nhân trước đây. Những người có thế chấp cao đã chọn vay. Khách hàng có mức sử dụng tín dụng trung bình hàng tháng cao hơn đã chọn vay. Khách hàng có thu nhập cao hơn có mức sử dụng thẻ tín dụng và thế chấp trung bình cao hơn. Cao học và Cao cấp/Chuyên nghiệp có mức sử dụng thẻ tín dụng hàng tháng cao hơn và đã vay ngân hàng.
